### 파생변수 만들기!
- 기존에 존재하는 변수를 이용해서 새로운 변수를 만들 수 있다!

-assign

```python
#새로운 total 변수를 만들기!
exam.assign(total=exam['math']+ exam['english']+ exam['science'],
           mean =(exam['math']+ exam['english']+ exam['science'])/3,)
```

```python
#파생변수 만들기 다른 방법
exam['total'] = exam['math']+ exam['english']+ exam['science']
```

```python
#np.where 할당하기!
import numpy as np
```

```python
exam.assign(test = np.where(exam['math']>=80, 'good','bad'))
```

```python
#새로운 total 변수를 만들기!
exam.assign(total=exam['math']+ exam['english']+ exam['science']).sort_values('total',ascending=False)
```

```python
#lambda 식을 이용해서 전처리!
exam.assign(new_total = lambda x: x['math']+x['english']+x['science'],
           mean = lambda x: x['new_total']/3) #lambda 식을 이용한 파생변수 만들기!
```

### 집단별로 요약하는 Groupby
- agg 더해서 통계치를 구하는 경우가 자주 발생!

- agg는 요약통계를 다 사용할 수 있다.
- mean, max, sum, min, std, median 등등

```python
exam.groupby('nclass').agg(mean_math = ('math','mean'))
#nclass 반을 그룹으로 나눈 후에 -> 수학이란 컬럼을 선쟁해서 컬럼에 대한 mean 평균을 계산해라!
#'math','mean' 작동되는 이유는 agg를 이용해서 적용이 된다!
```

```python
exam.groupby('nclass').agg(mean_math = ('math','mean'), #평균
                          sum_math =('math','sum'), #합
                          median_math=('math','median'), #중앙값
                          ct = ('math','count'))#빈도,개수
                        
```
