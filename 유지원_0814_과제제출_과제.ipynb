# 과제 

```python
mpg = pd.read_csv('mpg.csv')
```

### cty :도시 연비
### hwy : 고속도로 연비 
### fl : 연료
### displ : 배기량 
### cyl : 실린더 개수 

# 자동차 구매 의사 결정 분석 

## category로 분류해서 각 몇개씩 인지 세어주세요 

```python
mpg['category'].value_counts()
```

## suv인 차량을 카운트 해주세요 . 

```python
mpg.query("category=='suv'").groupby('manufacturer')['category'].count()
```

## fl(연료 ) 가 비싼 p는 제외시켜주세요 . 

```python
idx=mpg[mpg['fl']=='p'].index
mpg.drop(idx, inplace=True)
mpg
```

## 자동차 제조 회사 별로 모델, 카테고리, 년도로 묶고, (cyl,cty, hwy, displ )의 평균을 구해서 보기 편하게 만들어주세요.

```python
new_mpg=mpg.groupby(['manufacturer','model','category','year','fl']).agg(mean_cyl=('cyl','mean'),
                                                           mean_cty=('cty','mean'),
                                                           mean_hwy=('hwy','mean'),
                                                           mean_displ=('displ','mean'))
new_mpg
```

## 도시연비와 고속도로 연비를 합쳐주세요 . 

```python
new_mpg["mean_total"] = new_mpg['mean_cty'] + new_mpg['mean_hwy'] 
new_mpg
```

## mean_total 이 높은 상위 10개를 보여주세요 .

```python
new_mpg.sort_values('mean_total',ascending = False).head(10)
```

## mean_total 이 낮은 하위 10개를 보여주세요 .

```python
new_mpg.sort_values('mean_total').head(10)
```

### 연비가 좋을수록 displ( 배기량 ) 은 높아지고 , cyl (실린더)는 낮아진다.


## mean_total은 높고, mean_displ(배기량)은 낮고, cyl 가 적은 차량을 골라주세요 .

```python
new_mpg.sort_values(['mean_total','mean_displ', 'mean_cyl'], ascending = [False, True, True]) 
```
