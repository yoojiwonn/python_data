### 4. 각 나라별 학력 통계 계산하기

- pivot table을 이용하여 나라별 통계를 계산해봅시다.

#### pivot table 만들기

```python
countries = pd.pivot_table(data=survey.loc[1:, ["Q3", "Q4"]], index="Q3", columns="Q4", aggfunc= {"Q4" : "counts"}, fill_value=0) # Q4를 values값으로 넣고 
# NAN값에 0을 넣는다 
countries
```

#### 특정 국가 가져오기

```python
# 미국과 캐나다를 가져와봅시다.
plt.figure(figsize=(6, 12))
usa = countries.loc["United States of America"]
canada = countries.loc["Canada"]
plt.subplot(2,1,1)
plt.title("the highest level of formal education in USA", fontsize=14)
usa.plot(kind="barh")
plt.subplot(2,1,2)
plt.title("the highest level of formal education in Canada", fontsize=14)
canada.plot(kind="barh")
# display(usa)
# display(canada) # 두개다 출력 
```

#### 프로그래밍 언어 선호도를 포함하는 column들을 불러오기

```python
Q7_columns = survey.columns[survey.columns.str.startswith("Q7")] # Q7으로 시작하는 칼럼들을 뽑은것 
survey[["Q3"]+ list(Q7_columns)]
```

#### 언어별 정보가 합쳐진 DataFrame 만들기

```python
q7_list = []

for _, row in survey[Q7_columns][1:].iterrows(): # _, : series들만 출력이 가능하게 한것
  # print(row[~row.isnull()].values) #각 Row별로 응답한 정보만 리스트로 출력.
  q7_list.append(row[~row.isnull()].values)

q7_list[:2]
```

```python
survey["PL"] = ["PL"] + q7_list
survey
```

### countplot을 그리기

```python
# x축을 나이대 별로 정렬하여 countplot을 출력해보세요.
q7_list = []

for _, row in survey[Q7_columns][1:].iterrows(): # _, : series들만 출력이 가능하게 한것
  # print(row[~row.isnull()].values) #각 Row별로 응답한 정보만 리스트로 출력.
  q7_list.append(row[~row.isnull()].values)

q7_list
```

```python
from collections import Counter

q7_data = []

for row in q7_list:
  q7_data = q7_data + row

counter = Counter(q7_data)

df = pd.DataFrame({"Languages" : counter.key(), "Count": counter.values()})
plt.figure(figsize=(8,6))
sns.barplot(data=df, x="Count", y="Languages")
plt.show()
```
